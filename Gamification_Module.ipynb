#  Gamification Engine for Health Workers (MVP)

This project simulates a gamification system designed to motivate field-level health workers.

It includes points, streaks, and badge systems aligned with real-world healthcare workflows.

```python
workers = {
   "Anjali": {"points": 0, "streak": 0}
}

actions = {
   "visit": 10,
   "vaccination": 20,
   "registration": 15
}
```

```python
def perform_action(worker, action):
   workers[worker]["points"] += actions[action]
```

```python
def update_streak(worker, worked_today=True):
   if worked_today:
       workers[worker]["streak"] += 1
   else:
       workers[worker]["streak"] = 0
```

```python
def get_badge(points):
   if points >= 100:
       return "Health Champion"
   elif points >= 50:
       return "Rising Star"
   else:
       return "Beginner"
```

```python
perform_action("Anjali", "visit")
perform_action("Anjali", "vaccination")
update_streak("Anjali")

for w in workers:
   print("Worker:", w)
   print("Points:", workers[w]["points"])
   print("Streak:", workers[w]["streak"])
   print("Badge:", get_badge(workers[w]["points"]))
```


## Design Considerations
- Designed for offline-first environments
- Can be synced using background jobs (WorkManager equivalent)
- Modular logic for easy integration into mobile apps
